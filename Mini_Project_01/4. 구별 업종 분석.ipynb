#### [4.3.1] 데이터 분포 알아보기_1
* '자치구' 등 여러 변수를 기준으로 시각화 해보고 인사이트를 도출해보세요.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec

def draw_bar_graphs(dataframe, index_column=None, num_cols=2, figsize=(12, 8), gap=0.5):
    if index_column:
        dataframe = dataframe.set_index(index_column)

    numeric_columns = dataframe.select_dtypes(include=['number']).columns
    num_plots = len(numeric_columns)
    num_rows = int(np.ceil(num_plots / num_cols))

    # Create GridSpec for custom layout
    grid = GridSpec(num_rows, num_cols, wspace=0.5, hspace=gap)  # Adjust hspace parameter

    # Create subplots using GridSpec
    fig = plt.figure(figsize=figsize)
    axes = [fig.add_subplot(grid[i]) for i in range(num_plots)]

    # Loop through numeric columns and create bar graphs
    for ax, col in zip(axes, numeric_columns):
        dataframe[col].plot(kind='bar', ax=ax)
        ax.set_title(col)
        ax.set_xlabel(index_column if index_column else "Index")
        ax.set_ylabel("Value")
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right")  # Rotate x-axis labels for better readability

    # Hide any unused subplots
    for i in range(num_plots, num_rows * num_cols):
        axes[i].axis('off')

    plt.tight_layout()  # Adjust spacing between subplots
    plt.show()
```

```python
draw_bar_graphs(df.iloc[1:].drop(['기간'], axis=1), index_column='자치구')
```
